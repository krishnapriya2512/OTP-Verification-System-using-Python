### Data Cleaning

```python
# Removal of unnecessary columns - unnamed columns are only there for index
laptop_df.drop('Unnamed: 0', axis=1, inplace=True)
laptop_df.head()
```

### Checking the missing values

```python
missing_rows = laptop_df[laptop_df.isnull().any(axis=1)]
print(missing_rows.sample(10))
```

```python
# As there are no values in the entire rows of missing values, we can drop the rows
laptop_df.dropna(inplace= True)
```

```python
laptop_df.info()
```

```python
#Checking rows with "?" in the dataset
laptop_df[laptop_df.apply(lambda row: row.astype(str).str.contains(r'\?', regex=True).any(), axis=1)]
```

There are missing values in inches, memory and weight. Once we clean the data we will be inputing the values. 

```python
# Replacing '?' with Nan

laptop_df.replace('?', np.nan, inplace=True)
```

```python
# Removing 'GB' from Ram and convert to integer
laptop_df['Ram'] = laptop_df['Ram'].str.replace('GB', '').astype(int)
```

```python
laptop_df.head()
```

```python
laptop_df['Weight'] = laptop_df['Weight'].str.replace('kg', '').astype(float)
```

```python
laptop_df.head()
```

```python
laptop_df['Memory'].value_counts()
```

#### Before cleaning the other column I would like to impute the missing values for efficient data cleaning

```python
laptop_df.info()
```

```python
# Converting inches into float as it is in object datatype
laptop_df["Inches"] = laptop_df["Inches"].astype(float)
```

```python
# Imputing the missing values in Inches and Weights with median and using mode for Memory column
laptop_df['Inches'].fillna(laptop_df['Inches'].median(), inplace=True) 
laptop_df['Weight'].fillna(laptop_df['Weight'].median(), inplace=True)
laptop_df['Memory'].fillna(laptop_df['Memory'].mode()[0], inplace=True)
```

```python
# Checking if there are any null values
laptop_df.isnull().sum()
```

#### Resuming Data cleaning

```python
laptop_df["Memory"].unique()
```

```python
# To filter rows where memory contains "TB"
unique_tb_memories = laptop_df[
    laptop_df["Memory"].str.contains('TB', na=False)
].drop_duplicates(subset=["Memory"])

print(unique_tb_memories[["Memory"]])  # Shows only the unique 'memory' entries
```

```python
# Replace storage units to make all sizes in GB
laptop_df['Memory'] = laptop_df['Memory'].str.replace('1.0TB', '1024GB')
laptop_df['Memory'] = laptop_df['Memory'].str.replace('2.0TB', '2048GB')
laptop_df['Memory'] = laptop_df['Memory'].str.replace('1TB', '1024GB')
laptop_df['Memory'] = laptop_df['Memory'].str.replace('2TB', '2048GB')
```

```python
# Extract components using regex
extracted = laptop_df["Memory"].str.extractall(
    r'(?P<size>\d+\.?\d*)\s*(?P<unit>GB)\s*(?P<type>SSD|HDD|Flash Storage|Hybrid)'
).reset_index()

print(extracted.head())
```
